```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

data_frame= pd.read_csv("movie_dataset.csv")
```

```python
#using head i have printed 10 rows
data_frame.head(10)
```

```python
#take some columes of data which can help to find similarity 

dataset_features = ['keywords','cast','genres','director']
```

```python
#here we have join the features of daaset_features list created above code 

def join_dataset_features(row):
    return row['keywords']+" "+row['cast']+" "+row['genres']+" "+row['director']
```

```python
#  here we have used this code to filling all NaNs with blank string
for feature in dataset_features:
    data_frame[feature] = data_frame[feature].fillna('') 

#we have applied join_dataset_feature in all the rows and stored the result in combine_dataset_features

data_frame["combine_dataset_features"] = data_frame.apply(join_dataset_features,axis=1)
```

```python
#iloc[0] will print the first row of combine_datset_features

data_frame.iloc[0].combine_dataset_features
```

```python
def TitleFromIndex(index):
    return data_frame[data_frame.index == index]["title"].values[0]
def IndexFromTitle(title):
    return data_frame[data_frame.title == title]["index"].values[0]
```

```python
#creating new object of CountVectorizer()
CV = CountVectorizer() 

#feeding combined strings(movie contents) to CountVectorizer() object

count_matrix = CV.fit_transform(data_frame["combine_dataset_features"]) 
```

```python
cosine_similarity = cosine_similarity(count_matrix)
```

```python
Movie = "Harry Potter and the Half-Blood Prince"
movie_index = IndexFromTitle(Movie)
 #accessing the row corresponding to given movie to find all the similarity scores for that movie.
Similar_movies = list(enumerate(cosine_similarity[movie_index]))
```

```python
sortingOfSimilar_movies= sorted(Similar_movies,key=lambda x:x[1],reverse=True)[1:]
```

```python
#apply for loop to print the output 

i=0
print("10 similar movies to "+Movie+" are:\n")
for element in sortingOfSimilar_movies:
    print(TitleFromIndex(element[0]))
    i=i+1
    if i>10:
        break
```
